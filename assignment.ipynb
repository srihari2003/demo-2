```python
# Task 4:Check for the null values in 'df' and find the data type of the columns.
```

```python
import pandas as pd
import numpy as np
random_values = np.random.randn(4,4)
df = pd.DataFrame(random_values , columns=['F1','F2','F3','F4'])
null_values = df.isnull().any()
print("Nullvalues")
print(null_values)
type_of_the_data = df.dtypes
print("Datatype:")
print(type_of_the_data)
```

```python
#to get null values 
import pandas as pd
import numpy as np
random_values = np.random.randn(4,4)
random_values[1,2] = None
random_values[3,0] = None
df = pd.DataFrame(random_values , columns=['F1','F2','F3','F4'])
print(df)
null_values = df.isnull().any()
print("Nullvalues")
print(null_values)
type_of_the_data = df.dtypes
print("Datatype:")
print(type_of_the_data)
```

```python
#Task 5:Display the 'Random value 2' & 'Random value 3' columns with location method and index location method.
```

```python
import pandas as pd
import numpy as np
random_values = np.random.randn(4,4)
df = pd.DataFrame(random_values , columns=['F1','F2','F3','F4']) #F=feautures
df=df.rename(columns={'F1':'Random value 1','F2':'Random value 2','F3':'Random value 3','F4':'Random value 4'})
loc_col = df.loc[:,['Random value 2','Random value 3']]#using loc method
print("Loc method:")
print(loc_col)
index_col = df.iloc[:,[1,2]]#using index location method
print("Index method:")
print(index_col)
```
